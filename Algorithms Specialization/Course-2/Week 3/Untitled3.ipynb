```python
import numpy as np
import heapq
```

```python
def heapsort(arr):
    arr_sorted=[]
    heapq.heapify(arr)
    for i in range(len(arr)):
        arr_sorted.append(heapq.heappop(arr))
    return arr_sorted
    
```

```python
bv=[]
for i in range(len(arr)):
    sled_arr=heapsort(arr[:i+1])
    if((i+1)%2==0):
        bv.append(sled_arr[int(i/2)])
    else:
        
        bv.append(sled_arr[int((i+1)/2)])
        
    
    
    
```
